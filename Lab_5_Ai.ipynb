# **Case Study Task 2**

**Data Set Genteration**

```python
import numpy as np
import pandas as pd

# Set seed for reproducibility
np.random.seed(0)

# Generate dummy data
order_ids = np.arange(1, 501)
products = np.random.choice(['Product_A', 'Product_B', 'Product_C', 'Product_D', 'Product_E',
                             'Product_F', 'Product_G', 'Product_H', 'Product_I', 'Product_J'], size=500)
prices = np.random.uniform(10, 1000, size=500)
quantities = np.random.randint(1, 21, size=500)
dates_of_purchase = pd.to_datetime(np.random.choice(pd.date_range('2023-01-01', '2023-12-31'), size=500))

# Create DataFrame
order_data = pd.DataFrame({
    'Order ID': order_ids,
    'Product': products,
    'Price': prices,
    'Quantity': quantities,
    'Date of Purchase': dates_of_purchase
})

# Display the first few rows
order_data.head()
```

**Task 2 NumPay**

```python
price_quantity_array = order_data[['Price', 'Quantity']].to_numpy()

# Calculate total sales for each order (Price * Quantity)
total_sales = np.prod(price_quantity_array, axis=1)

# Add total sales as a new column in the DataFrame
order_data['Total Sales'] = total_sales

# Display the first few rows with Total Sales
order_data.head()
```

**Task 3 Pandas**

```python
high_value_orders = order_data[order_data['Total Sales'] > 100]

# Display filtered data
high_value_orders.head()
```

**Task 4 Pandas Group Product**

```python
total_quantity_by_product = order_data.groupby('Product')['Quantity'].sum()

# Display the total quantity sold for each product
total_quantity_by_product
```

**Task 5 Matplotlib**

```python
import matplotlib.pyplot as plt

# Create a scatter plot to visualize the relationship between Price and Quantity
plt.figure(figsize=(8, 6))
plt.scatter(order_data['Price'], order_data['Quantity'], color='blue', alpha=0.5)
plt.title('Scatter Plot of Price vs Quantity Sold')
plt.xlabel('Price ($)')
plt.ylabel('Quantity Sold')
plt.grid(True)
plt.show()
```

**Task 6: Matplotlib Task (Histogram of total sales values)**

```python
plt.figure(figsize=(8, 6))
plt.hist(order_data['Total Sales'], bins=20, color='green', edgecolor='black', alpha=0.7)
plt.title('Histogram of Total Sales')
plt.xlabel('Total Sales ($)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
